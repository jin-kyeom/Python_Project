```python
import requests

# 코드를 작성하세요
rating_pages = []
for i in range(2010, 2013):
    num_year = i
    for j in range(1,13):
        num_month = j
        for k in range(0,6):
            num_week = k
            Tmp_Address = "https://workey.codeit.kr/ratings/index?year=" + str(num_year) + "&month=" + str(num_month) + "&weekIndex=" + str(num_week)
            #print(Tmp_Address)
            response = requests.get(Tmp_Address)
            rating_pages.append(response.text)
# 테스트 코드
print(len(rating_pages)) # 가져온 총 페이지 수 
print(rating_pages[0]) # 첫 번째 페이지의 HTML 코드
```
